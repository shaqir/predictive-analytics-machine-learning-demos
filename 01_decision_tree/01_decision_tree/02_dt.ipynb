```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
import matplotlib.pyplot as plt

df = pd.read_csv('QCM7.csv', sep=';')

df.head()
```

```python
X = df.iloc[:,:10]
```

```python
class_cols = ['1-Octanol','1-Propanol','2-Butanol','2-propanol','1-isobutanol']
```

```python
y_str = df[class_cols].idxmax(axis=1)
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y_str)
y_encoded
```

```python
le.classes_ 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y_encoded, random_state=42, test_size=0.2)
```

```python
dt = DecisionTreeClassifier(max_depth=3, random_state=42)
dt
```

```python
dt.fit(X_train,y_train)
dt
```

```python
y_pred = dt.predict(X_test)
```

```python
print(f'Accuracy of Decision Tree is :{accuracy_score(y_test,y_pred)}')
```

```python
print(classification_report(
    y_test,
    y_pred,
    target_names=le.inverse_transform(sorted(set(y_test)))
))
```
