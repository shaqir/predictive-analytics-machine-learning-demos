```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
```

```python
X = np.array([[85,92], [70,88], [95,95], [60,80]])

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.preprocessing import StandardScaler

X = np.array([[170,65], [180,80], [160,55], [175,70]])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.preprocessing import RobustScaler

#X = np.array([[100,175,200,400,321,420,501,80000,1000]])
X = np.array([
    [100, 200],
    [150, 250],
    [120, 220],
    [10000, 300]  # Outlier in first column!
])
```

```python
scaler = RobustScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```
