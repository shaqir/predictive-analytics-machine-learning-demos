```python
import numpy as np
from sklearn.metrics.pairwise import cosine_distances
from scipy.spatial.distance import euclidean, cityblock 
```

```python
A = np.array([11,22,70,85,91])
B = np.array([69,44,75,36,500])
```

```python
eucl_dist = euclidean(A,B)
print("Euclidean distance:",eucl_dist)
```

```python
manhattan_dist =cityblock(A,B)
print("Manhattan distance:",manhattan_dist)
```

```python
cosine_dist = cosine_distances(A.reshape(1, -1), B.reshape(1, -1))[0][0]
print("Cosine distance:",cosine_dist)
```

```python
cosine_dist = cosine_distances([A],[B])[0][0]
print("Cosine distance:",cosine_dist)
```

```python
cosine_sim = 1 - cosine_dist
cosine_sim
```

```python
X = np.array([[1,0,3],[0,2,1],[1,1,1],[1,0,2]])
```

```python
dist_matrix_cosine = cosine_distances(X)
```

```python
print("dist_matrix_cosine: \n",dist_matrix_cosine)
```

```python
from sklearn.cluster import AgglomerativeClustering

clustering = AgglomerativeClustering(
    n_clusters= 2,
    metric='precomputed',
    linkage='average'
)
```

```python
labels = clustering.fit_predict(dist_matrix_cosine)
```
